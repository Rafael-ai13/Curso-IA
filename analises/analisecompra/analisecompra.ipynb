## Análise de compra

Os dados tem que ser equivalentes, as séries tem que ser equivalentes, mesma quantidade de séries, mesmo nome das séries, o tipo de cada série também deve ser o mesmo

```python
import pandas as pd
import matplotlib.pyplot as plt

# forma alternativa from matplotlit import matplot as plt

df_vendas1 = pd.read_csv('./dataset_clientes_completo.csv')
df_vendas2 = pd.read_csv('./dataset_clientes_completo_2.csv')
```

```python
# se não colocar esse ignore, ele vai duplicar os indices
df_completo = pd.concat([df_vendas1, df_vendas2], ignore_index=True)

df_completo
```

```python
# poucos valores nulos podem ser retirados
# outliers, colocar a  mediana nos valores nulos
# variaveis categoricos com valores nulos usar a moda
# distribuição normal, a gente vai aplicar a média
# normalização é passar todos para a mesma quantidade de casas decimais
# depois dá uma pesquisada melhor, para fins de relatar
```

```python
# filtrando valores, essa comparação te retorna um verdadeiro ou falso, tem que ter isso em mente

# df_completo['idade'] > 30
df_completo[df_completo['idade'] == 30] # mostra a tabela com base na condição, mostra só aqueles que cumprem as condições na tabela
```

```python
pessoas_MG = df_completo[(df_completo['cidade'] == 'MG') & (df_completo['idade'] == 30)]
pessoas_MG
```

```python
# isin é está em, vá até a série nome e procure os nomes que estão na listas, no caso Yuri e Igor, ele percorre toda a serie, procurando os valores da lista
df_completo[df_completo['nome'].isin(['Yuri', 'Igor'])]
```

```python
# encontrando valores nulos, mostre os registros onde a cidade- o campo cidade -  está nulo
df_completo[df_completo['cidade'].isna()]
```

```python
# comando oposto ao anterio, mostre todos os registro onde o campo cidade não está nulo
df_completo[df_completo['cidade'].notna()]
```
