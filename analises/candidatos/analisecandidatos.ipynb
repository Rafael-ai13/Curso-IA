```python
# vamos criar um dicionario para renomear as colunas com base no índice

renomear_por_indice = {0: 'id_candidato', 1:'data_cadastro', 2:'nome', 3:'genero', 4:'uf', 5:'idade', 6:'formacao', 7:'area_atuacao',
                       8:'experiencia_anos', 9:'certificacoes', 10:'idioma', 11:'salario_esperado', 12:'salario_vaga', 13:'disponibilidade_viagem',
                       14:'disponibilidade_mudanca', 15:'pcd', 16:'status_contratacao'}
```

```python
candidatos1.columns = [renomear_por_indice.get(i, col) for i, col in enumerate(candidatos1.columns)]
candidatos2.columns = [renomear_por_indice.get(i, col) for i, col in enumerate(candidatos2.columns)]
candidatos3.columns = [renomear_por_indice.get(i, col) for i, col in enumerate(candidatos3.columns)]

candidatos1.info()
```

```python
candidatos_completo = pd.concat([candidatos1, candidatos2, candidatos3], ignore_index=True)
candidatos_completo.info()
```

```python
# Verificando a quantidade de valores faltantes por coluna
valores_faltantes = candidatos_completo.isnull().sum()
print(valores_faltantes)
```

```python
candidatos_completo2 = candidatos_completo.copy()
```

```python
# excluir a coluna de nome
candidatos_completo2.drop('nome', axis=1, inplace=True)
candidatos_completo2
```

```python
# Lista de colunas categóricas para preencher com a moda
colunas_categoricas = ['data_cadastro','genero', 'uf', 'formacao', 'area_atuacao', 'idioma', 'disponibilidade_viagem', 'disponibilidade_mudanca', 'pcd']

# Lista de colunas numéricas para preencher com a mediana
colunas_numericas = ['experiencia_anos', 'certificacoes', 'salario_esperado', 'salario_vaga', 'status_contratacao', 'idade']

# Preenchendo as colunas categóricas com a moda
for col in colunas_categoricas:
    moda = candidatos_completo2[col].mode()[0]  # Calcula a moda da coluna
    candidatos_completo2[col] = candidatos_completo2[col].fillna(moda)  # Preenche os valores faltantes com a moda

# Preenchendo as colunas numéricas com a mediana
for col in colunas_numericas:
    mediana = candidatos_completo2[col].median()  # Calcula a mediana da coluna
    candidatos_completo2[col] = candidatos_completo2[col].fillna(mediana)  # Preenche os valores faltantes com a mediana

# Verificando se todos os valores faltantes foram preenchidos
print(candidatos_completo2.info())
```

```python
# Substituindo a coluna 'id' pelo valor do índice + 1
candidatos_completo2['id_candidato'] = candidatos_completo2.index + 1

# Verificando o resultado
print(candidatos_completo2[['id_candidato']].head())


```

## Análise dos candidatos contratados

Os candidatos contratados para vaga tinham predominantemente o nível de tecnólogo, as demais categorias não tiveram tanta diferença entre elas. Eles tiveram uma média de 5 anos de experiência, mais anos de experiência não significou mais chance de ganhar a vaga. A faixa de certificações foi de 1 a 3 certificações, as idades apresentam-se bem distribuídas com um pico na idade de 42 anos, a área de atuação mais contratado foi de tecnologia. A faixa de salário esperado foi de 6500 a 9000. Os salário das vagas e os salários esperados estão relativamente em conformidade, as distorções só se tornam grandes quando as pretensões salárias aumentam, há quebra de padrões vistos nas faixas anteriores. A faixa do salário da vaga foi de 4500 a 10000. Os contratados foram mais do sexo masculino e não PCD. A disponibilidade para mudança e a disponibilidade de viagem não foram decisivos, apresentam-se igualados, idioma idem, a uf que teve mais candidatos contratados foi a Bahia.
