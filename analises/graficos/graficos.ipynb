```python
import matplotlib.pyplot as plt  
```

```python
x = [1,2,3,4]
y = [2,3,4,5]
```

```python
# o mais correto é colocar o plt.show para mostrar
plt.plot(x, y, label ='dados')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('Grafico de linha')
plt.yticks([0,2,4,6])
plt.xticks([0,2,4,6])
plt.legend()
plt.show
```

```python
# lw é a grosura da linha e ls é a distância entre os meus pixels, é o pontinho, codigo hexadecimal para mostrar a cor
# pesquisar color picker no google lá terá a cor deseja e seus códigos em hex, rgb entre outros
plt.plot(x, y, ls=':', lw='10.0', color='#6b110b')
plt.show()
```

```python
# Grafico de barras

plt.bar(x, y, color='green')
plt.xlabel('quantidade')
plt.ylabel('quantidade')
plt.title('grafico de barras')
plt.show()
```

```python
# é o grafico de linha sem linha so mostra o ponto corresponte no eixo x e o no eixo y
plt.scatter(x, y, color='red', marker='v')
plt.show()
```

```python
# o primeiro numero e segundo numero são o tamanho de grafico em polegadas, isso em figuresize, no add_subplot os tres numeros devem ser lidos assim 2x2 vai dividir em 4 quadrados, o terceiro numero vai
# representar o index qual quadrado que ele é na ordem
figura = plt.figure(figsize=(20, 3))

figura.add_subplot(121)
plt.plot(x,y)
plt.title('grafico de linha')

figura.add_subplot(122)
plt.bar(x,y)
plt.title('grafico de barras')
```

```python
# dá um tipo de frequencia que o numero aparece. bom para graficos de vendas
valores = [3,5,10,3,1,8,7,7,7,10,5,3]
plt.hist(valores, label='frequencia')
plt.legend()
plt.show()
```
